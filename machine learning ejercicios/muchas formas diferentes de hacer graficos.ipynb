```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#array con los puntos
xpoints=np.array([5,34,37,56,78,99])
ypoints=np.array([73,89,32,35,65,62])
#plt.plot(xpoints,ypoints,'o')# la o hace que solo puntos, sin o seria linea
#plt.plot(xpoints,ypoints,marker='o',ms=20,mfc='r')# de lineas resaltando los puntos, tamano grande puntos
#interior puntos rojo
#plt.plot(xpoints,ypoints,'o:r')#linea discontinua roja
#plt.plot(xpoints,ypoints,linestyle='dotted')#discontinua
plt.plot(xpoints,ypoints,linestyle='dashed',color='g',linewidth='10')#discontinua mas fuerte, color verde, mas ancha
plt.show()
```

```python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

font1={'color':'blue','size':20}#defino tamano y color de fuente


plt.plot(x1, y1, x2, y2)#x con su y
plt.xlabel("Average Pulse")#eje x etiqueta
plt.ylabel("Calorie Burnage")#eje y etqieutea
plt.title("Sports Watch Data", fontdict=font1)#titulo. aisgno la fuente
plt.grid(color='green',linewidth='1')#la cuadricula de detras. verde y ancho de 1
plt.show()
```

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)#1 filas, 2 columna y es el primero
plt.plot(x,y)
plt.title("SALES")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)#1 filas, 2 columna y es el segundo
plt.plot(x,y)
plt.title("INCOME")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes,alpha=0.5)# de tamanos asigna este array, medio transparente

plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y, color = "red",width = 0.1)# de barras rojas, con tamano 0,1
#plt.barh(x,y)# de barras horizontales
plt.show()
```
