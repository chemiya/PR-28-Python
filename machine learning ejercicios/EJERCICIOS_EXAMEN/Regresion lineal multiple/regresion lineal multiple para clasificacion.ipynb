```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
wineData = pd.read_csv('winequality-red.csv')
wineData.head()#cargo datos
```

```python
wineData.quality.unique()#valores de la clase difentes
```

```python
wineData['category'] = np.where(wineData['quality'] >= 6, 1, 0)
wineData.head()#los mayores o iguales a 6 un 1 si no 0
```

```python
wineData = wineData.drop(['quality'], axis = 1)
wineData.head()#borro columna vieja
```

```python
from sklearn.model_selection import train_test_split

target = wineData['category'].copy()
print(target)#cojo columna de clase
```

```python
features = wineData.drop(['category'], axis=1)
print(features)#cojo columna de predictoras, eliminando la clase
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.2, random_state = 101)#separo
```

```python
print(X_train)
```

```python
X_train=X_train.reset_index(drop=True)#reinicio indices. otra vez en 0,1,2,3,4,5
X_test=X_test.reset_index(drop=True)
y_train=y_train.reset_index(drop=True)
y_test=y_test.reset_index(drop=True)
```

```python
print(X_train)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

#escalamos los datos
scaler.fit(X_train)

train_scaled = scaler.transform(X_train)
test_scaled = scaler.transform(X_test)

X_train = pd.DataFrame(train_scaled, columns = X_train.columns)
X_test = pd.DataFrame(test_scaled, columns = X_test.columns)

print(X_train)
```

```python
from sklearn.linear_model import LogisticRegression
 
logReg = LogisticRegression()#algoritmo
logReg.fit(X_train,y_train)#entrenamos
```

```python
predictions = logReg.predict(X_test)#hacemos prediccion
y_hat = pd.DataFrame(predictions, columns=["predicted"])
print(y_hat.head()) 
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_hat))#resultados 
```

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix

cf_matrix=confusion_matrix(y_test,y_hat)#matriz de confusion
ax= plt.subplot()
sns.heatmap(cf_matrix/np.sum(cf_matrix), annot=True,fmt='.2%', cmap='Blues', ax=ax);  

ax.set_xlabel('Predicted labels');
ax.set_ylabel('True labels'); 
ax.set_title('Confusion Matrix'); 
```

```python
wineData = pd.read_csv('winequality-red.csv')
wineData.head()#cargamos los datos
```

```python
c = []
for q in wineData['quality'].values:#recorremos columna, vamos poniendo valor numerico
    if q < 6:
        c.append(0)
    elif q > 6:
        c.append(2)
    else:
        c.append(1)

wineData['category'] = c
wineData.head()
```

```python
wineData = wineData.drop(['quality'], axis = 1)#eliminamos la antigua
wineData.head()
```

```python
from sklearn.model_selection import train_test_split

target = wineData['category'].copy()#cogemos la de la clase
features = wineData.drop(['category'], axis=1)#cogemos predictoras eliminando clase
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.1, random_state = 101)#separamos
```

```python
X_train=X_train.reset_index(drop=True)
X_test=X_test.reset_index(drop=True)
y_train=y_train.reset_index(drop=True)
y_test=y_test.reset_index(drop=True)#reiniciamos indices
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# escalamos los datos
scaler.fit(X_train)

train_scaled = scaler.transform(X_train)
test_scaled = scaler.transform(X_test)

X_train = pd.DataFrame(train_scaled, columns = X_train.columns)
X_test = pd.DataFrame(test_scaled, columns = X_test.columns)

X_train
```

```python
softReg = LogisticRegression(multi_class = 'multinomial', solver = 'lbfgs')#algoritmo
softReg.fit(X_train,y_train)#entrenamos
```

```python
predictions = softReg.predict(X_test)
y_hat = pd.DataFrame(predictions, columns=["predicted"])
print(y_hat.head(10)) #prediccion
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_hat))#resultados
```

```python
from sklearn.metrics import confusion_matrix#matriz de confusion

cf_matrix=confusion_matrix(y_test,y_hat)
ax= plt.subplot()
sns.heatmap(cf_matrix/np.sum(cf_matrix), annot=True,fmt='.2%', cmap='Blues', ax=ax);  

ax.set_xlabel('Predicted labels');
ax.set_ylabel('True labels'); 
ax.set_title('Confusion Matrix'); 
```
