```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#pd.leercsv
dataset=pd.read_csv("SocialNetworkAds.csv")
```

```python
print(dataset)
```

```python
x=dataset.iloc[:,[2,3]].values#age y salario
y=dataset.iloc[:,4].values#purchased
#se cogen sin etiquetas
#columna 2 y 3
```

```python
print("en x tengo:")
print(x)
print(x.shape)
print(type(x))
```

```python
print("en y tengo:")
print(y)
print(y.shape)
print(type(y))
```

```python
from sklearn.model_selection import train_test_split
```

```python
#separo en conjunto de entrenamiento y de test
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
print("para x_train:")
print(type(X_train))
print(len(X_train))
print(X_train.shape)
print("\n")
print("para x_test:")
print(type(X_test))
print(len(X_test))
print(X_test.shape)
print("\n")
print("para y_train:")
print(type(y_train))
print(len(y_train))
print(y_train.shape)
print("\n")
print("para y_test:")
print(type(y_test))
print(len(y_test))
print(y_test.shape)
print("\n")
print("para x:")
print(type(x))
print(len(x))
print(x.shape)
print("\n")
print("para y:")
print(type(y))
print(len(y))
print(y.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
#standarizar datos
```

```python
print("x_train antes:")
print(X_train)
X_train=sc.fit_transform(X_train)


```

```python
print("x_train despues:")
print(X_train)
```

```python
print("x_test antes:")
print(X_test)
X_test=sc.transform(X_test)
```

```python
print("x_test despues:")
print(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
#defino el algoritmo a utilizar

classifier=KNeighborsClassifier(n_neighbors=5,metric="minkowski",p=2)
#p=2 distancia euclidiana
```

```python
#entreno el modelo
classifier.fit(X_train,y_train)
```

```python
#hago una prediccion
y_pred=classifier.predict(X_test)
```

```python
print("y_test tengo:")
print(y_test)
```

```python
print("y_pred tengo:")
print(y_pred)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
#matriz de confusion
cm=confusion_matrix(y_test,y_pred)#ambos y
print(cm)
#diagonal principal bien 
#5 que eran no purchased han sido puestos como purchased
#         prediccion 
#          no  si
#real no       X
#     si
```
