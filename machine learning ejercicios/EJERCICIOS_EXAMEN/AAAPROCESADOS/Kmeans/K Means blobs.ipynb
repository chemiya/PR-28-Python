```python
import random 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans 
from sklearn.datasets import make_blobs 
%matplotlib inline
```

```python
np.random.seed(0)
```

```python
X, y = make_blobs(n_samples=5000, centers=[[4,4], [-2, -1], [2, -3], [1, 1]], cluster_std=0.9)
#ponemos los centros
```

```python
print(X)
```

```python
print(y)
```

```python
plt.scatter(X[:, 0], X[:, 1], marker='.',c=y)#dibujamos segun atributos, la clase iria en color
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=1/3,stratify=y)
```

```python
k_means = KMeans(init = "k-means++", n_clusters = 4, n_init = 12)
```

```python
k_means.fit(X_train,y_train)
```

```python
pred=k_means.predict(X_test)
pred
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,pred))
```

```python
k_means_labels = k_means.labels_
k_means_labels
```

```python
k_means_cluster_centers = k_means.cluster_centers_
z=k_means_cluster_centers
```

```python

plt.scatter(X_test[:,0],X_test[:,1],c=pred)
plt.scatter(z[:,0],z[:,1],c="black",marker="*")
plt.show()
```
