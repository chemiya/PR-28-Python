```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
df = load_iris()#cargo iris
```

```python

print(df) #imprimo todo
```

```python
X=df.data
print(X)#cojo solo los datos
```

```python
y=df.target
print(y)#cojo la clase
```

```python
#Separo los datos de "train" en entrenamiento y prueba para probar los algoritmos
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 50, test_size = 0.25)
#imprimos las formas que tienen
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
print(type(X_train))
print(type(X_test))
print(type(y_train))
print(type(y_test))
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
st=StandardScaler()
#z = (x - media) / devst.  media=0 y devst=1
```

```python
print("X_train antes:")
print(X_train)
```

```python
X_train=st.fit_transform(X_train)
```

```python
print("X_train despues:")
print(X_train)
```

```python
print("X_test antes:")
print(X_test)
```

```python
X_test=st.transform(X_test)
```

```python
print("X_test despues:")
print(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()#escojo el algoritmo
clf.fit(X_train,y_train)#lo entreno
```

```python
y_pred = clf.predict(X_test)# hago la prediccion
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score
sc=accuracy_score(y_pred,y_test)
print(sc)
```
